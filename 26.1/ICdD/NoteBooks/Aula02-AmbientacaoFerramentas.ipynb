## Revisão de Python: Controle de Fluxo

Python usa indentação para definir blocos de código em estruturas de controle:

````python
# Condicionais
nota_final = 7.5
if nota_final >= 7.0:
    print("Aprovado")
elif nota_final >= 5.0:
    print("Recuperação")
else:
    print("Reprovado")

# Loops
for i in range(5):
    print(f"Número: {i}")

notas = [8.5, 4.0, 9.5, 6.0]
aprovados = 0
for nota in notas:
    if nota >= 7.0:
        aprovados += 1
print(f"Total de aprovados: {aprovados} de {len(notas)}")
````

## Revisão de Python: Funções

As funções em Python são definidas usando a palavra-chave `def`:

````python
# Definindo uma função
def calcular_media(valores):
    if not valores:  # Lista vazia
        return 0
    return sum(valores) / len(valores)

# Usando a função
notas_turma_a = [8.5, 7.0, 9.5, 8.0]
media_turma_a = calcular_media(notas_turma_a)
print(f"Média da Turma A: {media_turma_a:.2f}")

# Função com parâmetros padrão
def classificar_aluno(nota, aprovacao=7.0):
    return "Aprovado" if nota >= aprovacao else "Reprovado"

print(classificar_aluno(8.5))        # Aprovado
print(classificar_aluno(6.5, 6.0))   # Aprovado (critério diferente)
````

## Introdução ao NumPy

NumPy é uma biblioteca fundamental para computação científica em Python:
- Trabalha com arrays multidimensionais de maneira eficiente
- Oferece funções matemáticas otimizadas
- É a base para muitas outras bibliotecas (pandas, scikit-learn)

````python
import numpy as np

# Criando arrays
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.zeros((3, 3))                # Matriz 3x3 de zeros
arr3 = np.random.rand(2, 4)            # Matriz 2x4 de valores aleatórios
arr4 = np.arange(0, 10, 2)             # [0, 2, 4, 6, 8]

print(f"Array 1: {arr1}")
print(f"Array 2:\n{arr2}")
print(f"Array 3:\n{arr3}")
print(f"Array 4: {arr4}")
print(f"Dimensões: {arr1.shape}, {arr2.shape}, {arr3.shape}")
print(f"Tipos: {arr1.dtype}, {arr3.dtype}")
````

## Operações com NumPy

Uma das grandes vantagens do NumPy é a vetorização das operações, tornando-as muito mais rápidas:

````python
import numpy as np

# Operações vetorizadas (muito mais rápidas)
arr = np.array([1, 2, 3, 4, 5])
print(f"Array original: {arr}")
print(f"Array * 2: {arr * 2}")          # [2, 4, 6, 8, 10] - Multiplicação elemento a elemento
print(f"Array ** 2: {arr ** 2}")        # [1, 4, 9, 16, 25] - Elevado ao quadrado
print(f"Raiz quadrada: {np.sqrt(arr)}")  # Raiz quadrada de cada elemento

# Operações entre arrays
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
print(f"\nArray 1: {arr1}")
print(f"Array 2: {arr2}")
print(f"Soma: {arr1 + arr2}")      # [6, 8, 10, 12]
print(f"Produto: {arr1 * arr2}")   # [5, 12, 21, 32] - Produto elemento a elemento
print(f"Produto escalar: {np.dot(arr1, arr2)}")  # 70 (1*5 + 2*6 + 3*7 + 4*8)
````

## Manipulação de Dados com Pandas

O Pandas oferece várias maneiras de selecionar e filtrar dados:

````python
import pandas as pd

# DataFrame de exemplo
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo'],
    'Idade': [22, 25, 23, 24, 21],
    'Departamento': ['TI', 'Marketing', 'TI', 'RH', 'Marketing'],
    'Salário': [3500, 4200, 3800, 3900, 3600]
}
df = pd.DataFrame(dados)

print("DataFrame original:")
print(df)

# Selecionando colunas
print("\nApenas coluna 'Nome':")
print(df['Nome'])  # Uma coluna (retorna Series)

print("\nColunas 'Nome' e 'Salário':")
print(df[['Nome', 'Salário']])  # Múltiplas colunas (retorna DataFrame)

# Filtrando linhas
print("\n* Funcionários com idade maior que 23:")
filtro=df['Idade'] > 23
print(" -> Filtro de dados para idade > 23")
print(filtro)
print(" -> Aplicando o filtro")
print(df[df['Idade'] > 23])  # Idades maiores que 23

print("\nFuncionários do departamento de TI:")
print(df[df['Departamento'] == 'TI'])  # Departamento de TI
````

## Funções Avançadas do Pandas

Vamos explorar agrupamento, ordenação e criação de novas colunas:

````python
import pandas as pd

# DataFrame de exemplo
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo'],
    'Idade': [22, 25, 23, 24, 21],
    'Departamento': ['TI', 'Marketing', 'TI', 'RH', 'Marketing'],
    'Salário': [3500, 4200, 3800, 3900, 3600]
}
df = pd.DataFrame(dados)

# Agrupamento
print("Média salarial por departamento:")
por_depto = df.groupby('Departamento')
print(por_depto['Salário'].mean())

# Ordenação
print("\nFuncionários ordenados por salário (maior para menor):")
print(df.sort_values('Salário', ascending=False))

# Criando novas colunas
df['Faixa Etária'] = ['20-22' if idade < 23 else '23-25' for idade in df['Idade']]
print("\nDataFrame com nova coluna 'Faixa Etária':")
print(df)
````

## Introdução à Visualização com Matplotlib

Matplotlib é a biblioteca fundamental para visualização em Python:
- Interface similar ao MATLAB
- Base para outras bibliotecas (seaborn, pandas plotting)
- Controle total sobre elementos gráficos
- Suporta múltiplos formatos de saída

````python
import matplotlib.pyplot as plt
import numpy as np

# Configuração para melhor visualização no Colab
plt.style.use('seaborn-v0_8-darkgrid')

# Dados de exemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = (x/10.0) - 0.25

# Criando um gráfico simples
plt.figure(figsize=(10, 4))
plt.plot(x, y1, label='sen(x)', color='blue', linewidth=2)
plt.plot(x, y2, label='x/10.0', color='red' , linewidth=1, linestyle='--')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Linha Simples')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

## Tipos Básicos de Gráficos

````python
# Criando múltiplos tipos de gráficos
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# 1. Scatter Plot
np.random.seed(42)
x_scatter = np.random.randn(50)
y_scatter = 2 * x_scatter + np.random.randn(50) * 0.5
axes[0, 0].scatter(x_scatter, y_scatter, alpha=0.6, c='purple')
axes[0, 0].set_title('Scatter Plot')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y')

# 2. Bar Plot
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [23, 45, 56, 78, 32]
axes[0, 1].bar(categorias, valores, color='skyblue', edgecolor='navy')
axes[0, 1].set_title('Gráfico de Barras')
axes[0, 1].set_ylabel('Valores')

# 3. Histogram
dados_hist = np.random.normal(100, 15, 1000)
axes[1, 0].hist(dados_hist, bins=30, color='green', alpha=0.7, edgecolor='black')
axes[1, 0].set_title('Histograma')
axes[1, 0].set_xlabel('Valores')
axes[1, 0].set_ylabel('Frequência')

# 4. Pie Chart
sizes = [30, 25, 20, 15, 10]
labels = ['Python', 'R', 'SQL', 'Julia', 'Outros']
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title('Gráfico de Pizza')

plt.tight_layout()
plt.show()
````

## Operações Avançadas com Pandas

Vamos explorar merge, join e pivot tables:

````python
import pandas as pd

# DataFrames para merge
df_funcionarios = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Diana'],
    'Depto_ID': [101, 102, 101, 103]
})

df_departamentos = pd.DataFrame({
    'Depto_ID': [101, 102, 103],
    'Departamento': ['TI', 'RH', 'Vendas']
})

# Merge (similar ao JOIN do SQL)
df_completo = pd.merge(df_funcionarios, df_departamentos, on='Depto_ID')
print("DataFrame após merge:")
print(df_completo)

# Pivot Table
df_vendas = pd.DataFrame({
    'Vendedor': ['Ana', 'Bruno', 'Ana', 'Bruno', 'Ana', 'Bruno','Ana'],
    'Produto': ['A', 'A', 'B', 'B', 'C', 'C','B'],
    'Vendas': [100, 150, 200, 180, 120, 140,77]
})

pivot = df_vendas.pivot_table(values='Vendas',
                               index='Vendedor',
                               columns='Produto',
                               aggfunc='sum')
print("\nPivot Table:")
print(pivot)
````

````python
# DataFrame com dados de texto
df_texto = pd.DataFrame({
    'Nome': ['  Ana Silva  ', 'BRUNO COSTA', 'carlos santos', 'Diana Oliveira'],
    'Email': ['ana@email.com', 'BRUNO@GMAIL.COM', 'carlos@yahoo.com', 'diana@outlook.com']
})

print("DataFrame original:")
print(df_texto)

# Limpeza e padronização
df_texto['Nome_Limpo'] = df_texto['Nome'].str.strip().str.title()
df_texto['Email_Limpo'] = df_texto['Email'].str.lower()
df_texto['Dominio'] = df_texto['Email_Limpo'].str.split('@').str[1]

print("\nDataFrame após tratamento:")
print(df_texto[['Nome_Limpo', 'Email_Limpo', 'Dominio']])
````

## Conceitos importantes:
- Seleção e filtragem de dados é uma das operações mais comuns
- Múltiplas maneiras de selecionar dados: notação de colchetes, `.loc`, `.iloc`
  - `.loc` é inclusivo no limite superior (0:2 inclui índices 0, 1 e 2)
  - `.iloc` é exclusivo no limite superior (0:2 inclui índices 0 e 1)
- Operações condicionais criam máscaras booleanas

````python
import pandas as pd

# DataFrame de exemplo
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo'],
    'Idade': [22, 25, 23, 24, 21],
    'Departamento': ['TI', 'Marketing', 'TI', 'RH', 'Marketing'],
    'Salário': [3500, 4200, 3800, 3900, 3600]
}
df = pd.DataFrame(dados)
print("\n",20*"=", "Dataframe")
print(df)

print("\n",20*"=", "Exemplo de uso do loc/iloc")
# Selecionando com loc (baseado em rótulos)
print(df.loc[0])                    # Primeira linha
print(df.loc[0:2, 'Nome':'Idade'])  # Linhas 0-2, colunas Nome até Idade

# Selecionando com iloc (baseado em posição)
print(df.iloc[0, 0])               # Valor na primeira linha, primeira coluna
print(df.iloc[0:2, 0:2])           # Primeiras 2 linhas, primeiras 2 colunas

# Filtragens complexas
alto_salario = df['Salário'] > 3600
print("\n",20*"=", "Vetor/Máscara alto_salario > 3600")
print(alto_salario)
ti_dept = df['Departamento'] == 'TI'
print("\n",20*"=", "Vetor/Máscara ti_dept == 'TI'")
print(ti_dept)
print("\n",20*"=", "Resultado de df[alto_salario & ti_dept]")
print(df[alto_salario & ti_dept])  # TI com salário > 3800
````
