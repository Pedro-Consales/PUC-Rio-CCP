## Medidas de Tendência Central

### Média, Mediana e Moda

> **Média**: Valor central que representa o "ponto de equilíbrio" dos dados. Calculada somando todos os valores e dividindo pelo número de observações. Sensível a valores extremos (outliers).
>
> **Mediana**: Valor que divide o conjunto de dados ao meio quando ordenado. 50% dos valores estão abaixo e 50% acima. Mais robusta a outliers que a média.
>
> **Moda**: Valor mais frequente no conjunto de dados. Útil para identificar o valor típico ou mais comum. Pode haver múltiplas modas ou nenhuma.

> *Outliers são valores atípicos que se desviam significativamente
> do padrão geral dos dados. Exploraremos este conceito em detalhes mais adiante.*


```python
# @title
# Selecionando apenas variáveis numéricas
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

print("MEDIDAS DE TENDÊNCIA CENTRAL")
print("=" * 40)

for feature in features:
    mean_val = iris[feature].mean()
    median_val = iris[feature].median()
    mode_val = iris[feature].mode()[0]  # Primeira moda

    print(f"\n{feature.replace('_', ' ').title()}:")
    print(f"  Média:   {mean_val:.3f}")
    print(f"  Mediana: {median_val:.3f}")
    print(f"  Moda:    {mode_val:.3f}")
```

```python
# @title
# Visualização das medidas de tendência central
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
fig.suptitle('Medidas de Tendência Central: Média vs Mediana vs Moda', fontsize=14, y=1.05)

for idx, feature in enumerate(features):
    ax = axes[idx]

    # Histograma
    ax.hist(iris[feature], bins=20, alpha=0.7, color='skyblue', edgecolor='black')

    # Linhas verticais para média, mediana e moda
    mean_val = iris[feature].mean()
    median_val = iris[feature].median()
    mode_val = iris[feature].mode()[0]

    ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Média: {mean_val:.2f}')
    ax.axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Mediana: {median_val:.2f}')
    ax.axvline(mode_val, color='orange', linestyle=':', linewidth=2, label=f'Moda: {mode_val:.2f}')

    ax.set_title(feature.replace('_', ' ').title())
    ax.set_xlabel('Valor (cm)')
    ax.set_ylabel('Frequência')
    ax.legend(fontsize=7, loc='upper right')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("📊 INSIGHTS VISUAIS:")
print("• Média ≈ Mediana: distribuições relativamente simétricas")
print("• Múltiplas modas: sugerem múltiplos grupos")
```

## Medidas de Dispersão

### Amplitude, Variância, Desvio Padrão e Coeficiente de Variação

> **Amplitude**: Diferença entre o maior e menor valor. Medida simples que indica o alcance total dos dados, mas muito sensível a outliers.
>
> **Variância**: Média dos quadrados dos desvios em relação à média. Mede o quão espalhados estão os dados, mas em unidades quadradas.
>
> **Desvio Padrão**: Raiz quadrada da variância. Indica o quanto os dados se desviam da média, na mesma unidade dos dados originais. Regra empírica: ~68% dos dados estão dentro de ±1 desvio padrão.
>
> **Coeficiente de Variação**: Razão entre desvio padrão e média (em %). Permite comparar variabilidade entre variáveis com diferentes escalas. CV < 10% indica baixa variabilidade.

```python
# @title
print("MEDIDAS DE DISPERSÃO")
print("=" * 40)

for feature in features:
    amplitude = iris[feature].max() - iris[feature].min()
    variancia = iris[feature].var()
    desvio_padrao = iris[feature].std()
    coef_variacao = (desvio_padrao / iris[feature].mean()) * 100

    print(f"\n{feature.replace('_', ' ').title()}:")
    print(f"  Amplitude:           {amplitude:.3f}")
    print(f"  Variância:           {variancia:.3f}")
    print(f"  Desvio Padrão:       {desvio_padrao:.3f}")
    print(f"  Coef. de Variação:   {coef_variacao:.1f}%")
```

```python
# @title
# Visualização do conceito de desvio padrão
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Desvio Padrão e Regra Empírica (68-95-99.7)', fontsize=14, y=1.02)

# "achata" os eixos para iterar facilmente
axes = axes.flatten()

for idx, feature in enumerate(features):
    ax = axes[idx]
    mean = iris[feature].mean()
    std = iris[feature].std()
    cv = (std/mean) * 100

    # Histograma
    counts, bins, patches = ax.hist(iris[feature], bins=25, alpha=0.5, color='gray', edgecolor='black')

    # Linha da média
    ax.axvline(mean, color='red', linewidth=2, label='Média', zorder=5)

    # Faixas de desvio padrão
    ax.axvspan(mean-std, mean+std, alpha=0.3, color='yellow', label='±1σ (68%)', zorder=3)
    ax.axvspan(mean-2*std, mean+2*std, alpha=0.2, color='orange', label='±2σ (95%)', zorder=2)
    ax.axvspan(mean-3*std, mean+3*std, alpha=0.1, color='red', label='±3σ (99.7%)', zorder=1)

    ax.set_title(f'{feature.replace("_", " ").title()}\nCV: {cv:.1f}%')
    ax.set_xlabel('Valor (cm)')
    ax.set_ylabel('Frequência')
    ax.legend(fontsize=10, loc='upper right')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("📊 INTERPRETAÇÃO DO DESVIO PADRÃO:")
print("• Amarelo (±1σ): Contém ~68% dos dados")
print("• Laranja (±2σ): Contém ~95% dos dados")
print("• Vermelho claro (±3σ): Contém ~99.7% dos dados")
print("• CV < 20%: Baixa a média variabilidade em todas as características")
```

## Coeficiente de Variação por Espécie

### Medida relativa de dispersão

> **Análise por grupo**: Permite comparar a variabilidade relativa entre diferentes grupos (espécies). Útil para identificar qual grupo tem dados mais homogêneos ou heterogêneos, independente da escala das medidas.

```python
# @title
# Calculando coeficiente de variação por espécie
print("COEFICIENTE DE VARIAÇÃO (%) POR ESPÉCIE")
print("=" * 45)

cv_df = pd.DataFrame(index=features)

for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    cv_values = {}

    for feature in features:
        mean = subset[feature].mean()
        std = subset[feature].std()
        cv = (std / mean) * 100  # CV em porcentagem
        cv_values[feature] = cv

    cv_df[species] = pd.Series(cv_values)

display(cv_df.round(2))

print("\nINTERPRETAÇÃO:")
print("- CV < 10%:  Baixa variabilidade")
print("- CV 10-20%: Média variabilidade")
print("- CV > 20%:  Alta variabilidade")
```

## Medidas de Posição

### Quartis e Percentis

> **Quartis (Q1, Q2, Q3)**: Dividem os dados ordenados em 4 partes iguais. Q1: 25% dos dados abaixo; Q2 (mediana): 50% abaixo; Q3: 75% abaixo. Úteis para entender a distribuição dos dados.
>
> **IQR (Intervalo Interquartil)**: Diferença entre Q3 e Q1. Contém 50% central dos dados. É uma medida robusta de dispersão, não afetada por valores extremos.
>
> **Detecção de Outliers com IQR**: O método IQR é amplamente usado para identificar valores atípicos:
> - **Limite inferior**: Q1 - 1.5 × IQR
> - **Limite superior**: Q3 + 1.5 × IQR  
> - Valores fora desses limites são considerados outliers
> - Exemplo: Se Q1=25, Q3=75, então IQR=50. Outliers seriam valores < -50 ou > 150
> - Este método é robusto pois usa a mediana (Q2) e quartis, não sendo afetado pelos próprios outliers
>
> **Percentis**: Generalização dos quartis. P10 significa que 10% dos dados estão abaixo desse valor. P90-P10 indica o alcance de 80% central dos dados.

```python
# @title
print("MEDIDAS DE POSIÇÃO")
print("=" * 40)

for feature in features:
    q1 = iris[feature].quantile(0.25)
    q2 = iris[feature].quantile(0.50)  # mediana
    q3 = iris[feature].quantile(0.75)
    iqr = q3 - q1

    p10 = iris[feature].quantile(0.10)
    p90 = iris[feature].quantile(0.90)

    print(f"\n{feature.replace('_', ' ').title()}:")
    print(f"  Q1 (25%):     {q1:.3f}")
    print(f"  Q2 (50%):     {q2:.3f}")
    print(f"  Q3 (75%):     {q3:.3f}")
    print(f"  IQR (Q3-Q1): {iqr:.3f}")
    print(f"  P10:          {p10:.3f}")
    print(f"  P90:          {p90:.3f}")
```

## Correlação entre Variáveis

### Matriz de Correlação de Pearson

> **Correlação de Pearson**: Mede a força e direção da relação linear entre duas variáveis. Varia de -1 a +1:
> - r ≈ 0: sem correlação linear
> - r > 0: correlação positiva (variam juntas)
> - r < 0: correlação negativa (variam em direções opostas)
> - |r| > 0.7: correlação forte

```python
# @title
# Calculando matriz de correlação
correlation_matrix = iris[features].corr()

print("MATRIZ DE CORRELAÇÃO (Pearson)")
print("=" * 40)
display(correlation_matrix.round(3))

# Identificando correlações mais fortes
print("\nCORRELAÇÕES MAIS FORTES:")
print("-" * 30)

# Extrair pares únicos (triangular superior)
import numpy as np
mask = np.triu(np.ones_like(correlation_matrix), k=1).astype(bool)
correlations = correlation_matrix.where(mask).stack().reset_index()
correlations.columns = ['Var1', 'Var2', 'Correlacao']
correlations = correlations.sort_values('Correlacao', key=abs, ascending=False)

for _, row in correlations.head(5).iterrows():
    print(f"{row['Var1']} vs {row['Var2']}: {row['Correlacao']:.3f}")
```

```python
# @title
# Visualização de diferentes níveis de correlação
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Visualizando Correlações: Do Forte ao Fraco', fontsize=14, y=1.02)

# Exemplos de correlações
correlations_to_plot = [
    ('petal_length', 'petal_width', 'Correlação Muito Forte'),
    ('sepal_length', 'petal_length', 'Correlação Forte'),
    ('sepal_length', 'petal_width', 'Correlação Moderada-Forte'),
    ('petal_length', 'sepal_width', 'Correlação Moderada'),
    ('sepal_length', 'sepal_width', 'Correlação Fraca'),
    ('sepal_width', 'petal_width', 'Correlação Fraca-Moderada')
]

for idx, (x_var, y_var, title) in enumerate(correlations_to_plot):
    row = idx // 3
    col = idx % 3
    ax = axes[row, col]

    # Calcular correlação
    corr_value = iris[x_var].corr(iris[y_var])

    # Scatter plot com cores por espécie
    for species in iris['species'].unique():
        subset = iris[iris['species'] == species]
        ax.scatter(subset[x_var], subset[y_var], alpha=0.6, s=30, label=species)

    # Linha de tendência geral
    z = np.polyfit(iris[x_var], iris[y_var], 1)
    p = np.poly1d(z)
    x_line = np.linspace(iris[x_var].min(), iris[x_var].max(), 100)
    ax.plot(x_line, p(x_line), "r--", alpha=0.8, linewidth=2)

    ax.set_xlabel(x_var.replace('_', ' ').title())
    ax.set_ylabel(y_var.replace('_', ' ').title())
    ax.set_title(f'{title}\nr = {corr_value:.3f}')
    ax.grid(True, alpha=0.3)

    # Colorir o fundo baseado na força da correlação
    if abs(corr_value) > 0.8:
        ax.set_facecolor('#ffe6e6')  # Vermelho claro para forte
    elif abs(corr_value) > 0.5:
        ax.set_facecolor('#fff9e6')  # Amarelo claro para moderado
    else:
        ax.set_facecolor('#e6f3ff')  # Azul claro para fraco

# Adicionar legenda geral
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right', bbox_to_anchor=(1.05, 0.98))

plt.tight_layout()
plt.show()

print("📊 INTERPRETAÇÃO DAS CORRELAÇÕES:")
print("• r > 0.8: Correlação muito forte (fundo vermelho claro)")
print("• 0.5 < r < 0.8: Correlação moderada a forte (fundo amarelo claro)")
print("• r < 0.5: Correlação fraca (fundo azul claro)")
print("• Linha vermelha tracejada: Tendência linear geral")
print("• Cores dos pontos: Diferentes espécies de íris")
```

## Teste de Normalidade

### Shapiro-Wilk Test

> **Teste de Shapiro-Wilk**: Testa se os dados seguem uma distribuição normal (gaussiana). Importante para validar pressupostos de muitos testes estatísticos:
> - p-valor > 0.05: dados parecem normais
> - p-valor < 0.05: evidência contra normalidade
> - Útil para amostras pequenas (n < 50)

```python
# @title
# Teste de Shapiro-Wilk para normalidade
print("TESTE DE NORMALIDADE (Shapiro-Wilk)")
print("=" * 45)
print("H0: Os dados seguem uma distribuição normal")
print("H1: Os dados NÃO seguem uma distribuição normal")
print("\nNível de significância: α = 0.05")
print("-" * 45)

alpha = 0.05

for feature in features:
    stat, p_value = stats.shapiro(iris[feature])

    if p_value > alpha:
        conclusion = "✓ Normal (falha em rejeitar H0)"
    else:
        conclusion = "✗ Não-normal (rejeita H0)"

    print(f"\n{feature.replace('_', ' ').title()}:")
    print(f"  Estatística: {stat:.4f}")
    print(f"  p-valor:     {p_value:.6f}")
    print(f"  Conclusão:   {conclusion}")
```

```python
# @title
# Visualização de normalidade com Q-Q plots
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
fig.suptitle('Análise Visual de Normalidade: Q-Q Plots e Histogramas', fontsize=14, y=1.02)

for idx, feature in enumerate(features):
    # Q-Q plot
    ax1 = axes[0, idx]
    stats.probplot(iris[feature], dist="norm", plot=ax1)
    ax1.set_title(f'{feature.replace("_", " ").title()}')
    ax1.grid(True, alpha=0.3)

    # Obter p-valor do teste Shapiro-Wilk
    _, p_value = stats.shapiro(iris[feature])

    # Adicionar interpretação no título
    if p_value > 0.05:
        ax1.set_xlabel(f'p={p_value:.4f} (Normal)', fontsize=10)
        ax1.get_lines()[0].set_color('green')
    else:
        ax1.set_xlabel(f'p={p_value:.4f} (Não Normal)', fontsize=10)
        ax1.get_lines()[0].set_color('red')

    # Histograma com curva normal teórica
    ax2 = axes[1, idx]

    # Histograma dos dados reais
    n, bins, patches = ax2.hist(iris[feature], bins=15, density=True,
                                alpha=0.7, color='skyblue', edgecolor='black')

    # Curva normal teórica
    mu = iris[feature].mean()
    sigma = iris[feature].std()
    x = np.linspace(iris[feature].min(), iris[feature].max(), 100)
    ax2.plot(x, stats.norm.pdf(x, mu, sigma), 'r-', linewidth=2,
            label='Normal teórica')

    ax2.set_xlabel('Valor (cm)')
    ax2.set_ylabel('Densidade')
    ax2.legend(fontsize=10)
    ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("📊 INTERPRETAÇÃO DOS Q-Q PLOTS:")
print("• Pontos na linha diagonal: Dados seguem distribuição normal")
print("• Desvios nas caudas: Indicam caudas pesadas ou leves")
print("• Curvas em S: Indicam assimetria (skewness)")
print("• Múltiplos padrões: Sugerem múltiplas populações (espécies)")
print("\nOBSERVAÇÃO: Características das pétalas mostram clara não-normalidade")
print("devido à mistura de três espécies com distribuições distintas.")

print("\n💡 INSIGHT IMPORTANTE PARA PRÓXIMA ANÁLISE:")
print("Como as pétalas NÃO são normais, o teste ANOVA pode não ser ideal!")
print("• ANOVA assume normalidade dos dados")
print("• Testes não-paramétricos existem como alternativas")
print("• Na prática: ANOVA é robusto a desvios moderados de normalidade")
```

## ANOVA - Comparação entre Espécies

### Análise de Variância

> **ANOVA (Análise de Variância)**: É um teste estatístico usado para verificar se três ou mais grupos independentes têm médias significativamente diferentes:
> - H0: todas as médias são iguais
> - H1: pelo menos uma média é diferente
> - p-valor < 0.05: existe diferença significativa entre grupos  

> O ANOVA justamente compara:  
> - **Variabilidade entre os grupos** (diferenças nas médias)  
> - **Variabilidade dentro dos grupos** (ruído interno)  
> - *Requer homogeneidade de variâncias e normalidade*
>

> **Por que olhar a variabilidade entre as médias dos grupos?**  
> Imagine que a diferença entre as médias seja de apenas 2 pontos.  
> - Se dentro de cada grupo os valores variam muito (alta dispersão), essa diferença de 2 pode ser apenas **ruído aleatório**.  
> - Mas se dentro de cada grupo os valores são bem concentrados (baixa dispersão), então uma diferença de 2 já é **um sinal consistente de efeito real**.  
>

> Insight: só vale a pena dizer que os grupos são diferentes quando a **diferença entre médias é maior que a variação natural dentro de cada grupo**.




```python
# @title
# ANOVA para verificar diferenças entre espécies
print("ANOVA - ANÁLISE DE VARIÂNCIA")
print("=" * 40)
print("H0: Não há diferença entre as médias das espécies")
print("H1: Pelo menos uma espécie tem média diferente")
print("\nNível de significância: α = 0.05")
print("-" * 40)

for feature in features:
    # Separando os dados por espécie
    group1 = iris[iris['species'] == 'setosa'][feature]
    group2 = iris[iris['species'] == 'versicolor'][feature]
    group3 = iris[iris['species'] == 'virginica'][feature]

    # Realizando ANOVA
    f_stat, p_value = stats.f_oneway(group1, group2, group3)

    if p_value < alpha:
        conclusion = "✓ Há diferença significativa (rejeita H0)"
    else:
        conclusion = "✗ Sem diferença significativa (falha em rejeitar H0)"

    print(f"\n{feature.replace('_', ' ').title()}:")
    print(f"  F-estatística: {f_stat:.4f}")
    print(f"  p-valor:       {p_value:.8f}")
    print(f"  Conclusão:     {conclusion}")
```

```python
# @title
# Visualização do conceito de ANOVA
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
fig.suptitle('ANOVA Visual: Comparando Variâncias Entre e Dentro dos Grupos', fontsize=14, y=1.02)

for idx, feature in enumerate(features):
    # Violin plot para mostrar distribuições por grupo
    ax1 = axes[0, idx]
    parts = ax1.violinplot([iris[iris['species'] == s][feature].values
                            for s in iris['species'].unique()],
                           positions=[0, 1, 2],
                           widths=0.7,
                           showmeans=True,
                           showmedians=True)

    # Customizar cores
    colors = ['lightblue', 'lightgreen', 'lightcoral']
    for pc, color in zip(parts['bodies'], colors):
        pc.set_facecolor(color)
        pc.set_alpha(0.7)

    # Adicionar médias como pontos vermelhos
    for i, species in enumerate(iris['species'].unique()):
        mean_val = iris[iris['species'] == species][feature].mean()
        ax1.plot(i, mean_val, 'ro', markersize=8)
        ax1.text(i, mean_val + 0.1, f'{mean_val:.2f}', ha='center', fontsize=8)

    ax1.set_xticks([0, 1, 2])
    ax1.set_xticklabels(['Setosa', 'Versicolor', 'Virginica'], rotation=45)
    ax1.set_title(feature.replace('_', ' ').title())
    ax1.set_ylabel('Valor (cm)')
    ax1.grid(True, alpha=0.3, axis='y')

    # Gráfico de barras mostrando variâncias
    ax2 = axes[1, idx]

    # Calcular variâncias
    var_between = iris.groupby('species')[feature].mean().var()
    var_within = iris.groupby('species')[feature].var().mean()
    f_statistic = var_between / var_within * (3-1) * (150/(3-1))  # Aproximação do F

    bars = ax2.bar(['Entre\ngrupos', 'Dentro dos\ngrupos'],
                   [var_between, var_within],
                   color=['#ff9999', '#66b3ff'])

    # Adicionar valores nas barras
    for bar, val in zip(bars, [var_between, var_within]):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height,
                f'{val:.3f}', ha='center', va='bottom', fontsize=9)

    # Adicionar F-statistic no título
    _, p_value = stats.f_oneway(
        iris[iris['species'] == 'setosa'][feature],
        iris[iris['species'] == 'versicolor'][feature],
        iris[iris['species'] == 'virginica'][feature]
    )

    ax2.set_title(f'F = {f_statistic:.2f}, p = {p_value:.4f}')
    ax2.set_ylabel('Variância')
    ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("📊 COMPREENDENDO ANOVA VISUALMENTE:")
print("• Violins mostram a distribuição completa de cada grupo")
print("• Pontos vermelhos: médias de cada grupo")
print("• Quanto mais separadas as médias, maior a variância ENTRE grupos")
print("• Quanto mais estreitos os violins, menor a variância DENTRO dos grupos")
print("• F-statistic alto = médias são significativamente diferentes")
print("• p < 0.05 = rejeita H0 (as médias são diferentes)")
```

## Intervalos de Confiança

### IC 95% para as médias por espécie

> **Por que Intervalos de Confiança são importantes?**
>
> Imagine que você mediu o comprimento médio de pétalas de 50 flores setosa e encontrou 1.46 cm.
> Mas e se tivesse medido outras 50 flores? Provavelmente encontraria um valor ligeiramente diferente.
>
> O **Intervalo de Confiança** quantifica essa incerteza:
> - IC 95% = [1.41, 1.51] significa que se repetíssemos o experimento muitas vezes, 95% dos intervalos calculados conteriam a verdadeira média populacional
> - **Não é** "95% de chance da média estar neste intervalo" (interpretação comum mas incorreta)
> - **É** "este método produz intervalos corretos em 95% das vezes"
>
> **Usos práticos**:
> - Comparar grupos: ICs que não se sobrepõem sugerem diferença significativa
> - Avaliar precisão: IC estreito = estimativa precisa, IC largo = muita incerteza
> - Planejar amostras: ICs muito largos indicam necessidade de mais dados

>
  > **Erro Padrão (SEM)**: Estimativa do desvio padrão da distribuição amostral da média.
  > - SEM = desvio padrão / √n
  > - **Usado para calcular o IC**: IC = média ± (valor crítico × SEM)
  > - Menor SEM => IC mais estreito => estimativa mais precisa





  > **Como o IC é calculado?**
  >
  > IC = média ± (valor crítico × SEM)
  >
  > O **valor crítico** determina "quantos erros padrão" incluir no intervalo:
  > - Para IC 95%: valor crítico ≈ 2 (exatamente 1.96 para distribuição normal)
  > - Significa: pegamos 2 SEMs para cada lado da média
  > - Por que 2? Porque captura ~95% da distribuição normal
  > - Para amostras pequenas: usa-se a distribuição t (valor um pouco maior que 2)
  >
  > **Exemplo prático**:
  > - Média = 5.0, SEM = 0.1
  > - IC 95% = 5.0 ± (2 × 0.1) = [4.8, 5.2]
  > - Interpretação: estamos "2 erros padrão" confiantes

```python
# @title
# Calculando intervalos de confiança para as médias (95%)
print("INTERVALOS DE CONFIANÇA (95%) PARA AS MÉDIAS")
print("=" * 55)

confidence_intervals = pd.DataFrame(index=features)
species = iris['species'].unique()

for specie in species:
    subset = iris[iris['species'] == specie]
    ci_values = {}

    for feature in features:
        mean = subset[feature].mean()
        std_err = stats.sem(subset[feature])
        ci = stats.t.interval(0.95, len(subset[feature])-1, loc=mean, scale=std_err)
        ci_values[feature] = f"{mean:.2f} ({ci[0]:.2f}, {ci[1]:.2f})"

    confidence_intervals[specie] = pd.Series(ci_values)

display(confidence_intervals)
```

## O que são Outliers?

### Definição e Importância

> **Outliers (Valores Atípicos)**: São observações que se desviam significativamente do padrão geral dos dados. Podem indicar:
> - Erros de medição ou digitação
> - Eventos raros mas genuínos
> - Populações diferentes misturadas
> - Descobertas importantes

> **Por que detectar outliers?**
> - Podem distorcer análises estatísticas (média, desvio padrão)
> - Afetam modelos de machine learning
> - Podem revelar insights valiosos sobre o processo estudado
> - Ajudam a identificar problemas na coleta de dados

> **Métodos principais de detecção:**
> 1. **Método IQR**: Baseado em quartis (robusto, não-paramétrico)
> 2. **Z-Score**: Baseado em desvios padrão (assume normalidade)
> 3. **Visualização**: Boxplots, scatter plots (identificação visual)

> 💡 **Importante**: Nem todo outlier deve ser removido! Primeiro entenda sua origem.

## Z-Score e Detecção de Outliers

### Padronização e identificação de outliers

> **Z-Score**: Número de desvios padrão que um valor está distante da média:
> - Z = (x - μ) / σ
> - |Z| > 3: possível outlier (regra empírica)
> - |Z| > 2: valor incomum (95% dos dados estão dentro de ±2σ)
> - Útil para comparar valores de diferentes escalas

```python
# @title
# Calculando Z-scores para identificar outliers
print("DETECÇÃO DE OUTLIERS COM Z-SCORE")
print("=" * 40)
print("Critério: |Z-score| > 2.5 = outlier potencial")
print("-" * 40)

outliers_found = 0
outlier_details = []

for feature in features:
    # Calcular Z-scores
    mean_val = iris[feature].mean()
    std_val = iris[feature].std()
    z_scores = (iris[feature] - mean_val) / std_val

    # Identificar outliers (|Z| > 2.5)
    outliers = iris[np.abs(z_scores) > 2.5]

    print(f"\n{feature.replace('_', ' ').title()}:")
    if len(outliers) > 0:
        print(f"  Outliers encontrados: {len(outliers)}")
        for idx, row in outliers.iterrows():
            z_val = z_scores.iloc[idx]
            print(f"    Índice {idx}: {row[feature]:.2f} cm (Z={z_val:.2f}) - {row['species']}")
            outlier_details.append((feature, idx, row[feature], z_val, row['species']))
        outliers_found += len(outliers)
    else:
        print(f"  Nenhum outlier encontrado")

print(f"\n📊 RESUMO:")
print(f"Total de outliers: {outliers_found}")
if outliers_found > 0:
    species_outliers = {}
    for _, _, _, _, species in outlier_details:
        species_outliers[species] = species_outliers.get(species, 0) + 1
    print("Outliers por espécie:")
    for species, count in species_outliers.items():
        print(f"  {species}: {count} outliers")
```

## 🎯 Micro-atividade: Outlier Detective (8 minutos)

**Objetivo**: Comparar dois métodos de detecção de outliers e entender suas diferenças!

### Instruções:

1. **Execute a célula abaixo** que já contém a implementação do método IQR
2. **Compare os resultados** com o método Z-Score executado anteriormente
3. **Responda as perguntas** de análise no final da célula

### O que você vai aprender:
- Como diferentes métodos podem dar resultados diferentes
- Quando usar cada método
- A importância de entender o contexto antes de tratar outliers

### Critérios de avaliação:
- ✅ Execução correta do código IQR
- ✅ Comparação quantitativa dos métodos
- ✅ Reflexão sobre qual método é mais apropriado

*💡 Dica: O método IQR é mais robusto (não assume normalidade), enquanto o Z-Score é mais sensível a distribuições não-normais*

```python
# PARTE 1: Implementar método IQR
print("=" * 50)
print("MÉTODO IQR - DETECÇÃO DE OUTLIERS")
print("=" * 50)

# Contadores para comparação
outliers_iqr_total = 0

# IQR para cada característica
for feature in features:
    # Calcule Q1, Q3 e IQR
    q1 = iris[feature].quantile(0.25)
    q3 = iris[feature].quantile(0.75)
    iqr = q3 - q1

    # Calcule os limites
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Identifique outliers
    outliers_iqr = iris[(iris[feature] < lower_bound) | (iris[feature] > upper_bound)]
    outliers_iqr_total += len(outliers_iqr)

    # Mostre resultados
    print(f"\n{feature.replace('_', ' ').title()}:")
    print(f"  Q1 = {q1:.3f}, Q3 = {q3:.3f}, IQR = {iqr:.3f}")
    print(f"  Limites: [{lower_bound:.3f}, {upper_bound:.3f}]")
    print(f"  Outliers encontrados: {len(outliers_iqr)}")
    if len(outliers_iqr) > 0:
        print(f"  Valores outliers: {outliers_iqr[feature].values}")

# PARTE 2: Comparação dos Métodos
print("\n" + "=" * 50)
print("ANÁLISE COMPARATIVA DOS MÉTODOS")
print("=" * 50)

print("\n📊 RESULTADOS:")
print(f"• Método Z-Score detectou: {outliers_found} outliers no total")
print(f"• Método IQR detectou: {outliers_iqr_total} outliers no total")

print("\n🔍 OBSERVAÇÕES:")
if outliers_iqr_total > outliers_found:
    print("• O método IQR detectou MAIS outliers que o Z-Score")
    print("• Isso sugere que os dados têm distribuição com caudas leves")
elif outliers_iqr_total < outliers_found:
    print("• O método Z-Score detectou MAIS outliers que o IQR")
    print("• Isso ocorre porque Z-Score assume normalidade, que não temos nas pétalas")
else:
    print("• Ambos os métodos detectaram a mesma quantidade de outliers")

print("\n💡 CONCLUSÃO:")
print("• IQR é mais apropriado para este dataset porque:")
print("  - Não assume normalidade (vimos que pétalas não são normais)")
print("  - É robusto a distribuições assimétricas")
print("  - Baseado em mediana/quartis, não afetado por extremos")

print("\n📝 QUANDO USAR CADA MÉTODO:")
print("• Z-Score: Quando os dados são aproximadamente normais")
print("• IQR: Quando há assimetria, múltiplas modas ou distribuições não-normais")
```
